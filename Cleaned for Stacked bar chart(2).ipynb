```python
patientinfo.dropna(subset=['age'],inplace=True)
patientinfo.age.isnull().any()
needed = patientinfo[['sex','age']]
needed['age'] = needed['age'].apply(lambda x: int(x.strip('s')))
male = needed[needed['sex'] == 'male']
female = needed[needed['sex'] == 'female']
male = male.groupby("age").count().reset_index().rename(columns={"sex": "count"})
male["sex"] = 'male'
#male['age'] = male['age'].apply(lambda x: int(x.strip('s')))
female = female.groupby("age").count().reset_index().rename(columns={"sex": "count"})
female["sex"] = 'female'
final = pd.concat([male,female]).reset_index()[['age','count','sex']]
final
```
